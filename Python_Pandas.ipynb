```python
for i in students_performance['lunch'].unique():
    print('Дисперсия для группы с ',i,' ланчем равна: ', \
          round(students_performance[students_performance['lunch']==i][['math score','reading score','writing score']].values.std()**2,2), \
          ' , а среднее: ', \
          round(students_performance[students_performance['lunch']==i][['math score','reading score','writing score']].values.mean(),2))
    print(students_performance[students_performance['lunch']==i].describe())
```

```python
students_performance = students_performance\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})
```

```python
students_performance.head(5)
```

```python
mmm = 98
students_performance.query("math_score > @mmm & gender == 'male'")
```

```python
score_columns = [i for i in list(students_performance) if 'score' in i]
score_columns
```

```python
students_performance.filter(like='score')
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity'])\
    .aggregate({'math_score': 'mean','reading_score' : 'mean'})\
    .rename(columns = {'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score'})
```

```python
mean_scores.loc[[('female', 'group A'), ('male', 'group C')]]
```

```python
students_performance.sort_values(['gender', 'math_score'], ascending=False)\
.groupby('gender').head()
```

```python
students_performance['total_score'] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```
